```python
# import pandas
import pandas as pd
```

```python
#Read data
df = pd.read_csv("/Users/tien/Desktop/data.csv")
df.head()
```

```python
#remove rows with null values
df = df[df["y-kien"].notnull()]

df.head()
```

```python
df_drop = df.drop(['STT', 'Mã SV', 'Họ và tên'], axis=1)
df_drop
```

```python
df_drop.shape[0]
```

```python
num_columns = df_drop.columns.shape[0]
col_names = df_drop.columns.values.tolist()
df_drop.loc[:, 'merged'] = ""
for i in range(num_columns):
    print(col_names[i])
    df_drop.loc[:, 'merged'] = df_drop.loc[:, 'merged'] + df_drop[col_names[i]]
```

```python
list_ = df_drop.loc[:, 'merged']
list_
```

```python
type(list_)
```

```python
df_trans = pd.DataFrame(list_).reset_index()
df_trans.head()
```

```python
df_allMerged = df_trans.groupby("merged")["merged"].apply(' '.join)
df_allMerged
```

```python
from string import punctuation
from collections import Counter
s = str(df_allMerged)

words = s.translate(punctuation).lower().split()
word_count = Counter(words)
word_count.most_common(5)
```

```python
dict_X = dict(word_count)
dict_X.
```

```python
def findkey(x):
    for key, value in dict_X.items():
        if value == x:
            return key
```
